```python
#import pandas
import pandas as pd

#import raw data
df = pd.read_csv('../data/raw_dataset.csv')

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
#df.columns - Gets all the column names of the DataFrame.
#str.strip() - Removes whitespace from the begining and the end
#str.lower() - Converts the column names to lowercase
#str.replace(" ", "_") - Replaces any spaces in the column names with underscores.


#check data
print(df.head())
```

```python
#Remove duplicates 
df = df.drop_duplicates()
print(df.shape)

#Check for missing values
print(df.isnull().sum())
```

```python
#Behavioural columns 
zero_fill_cols = ['social_media_engagement_score',
                  'mobile_app_usage',
                  'wishlist_items',
                  'product_reviews_written',
                  'customer_service_calls',
                  'session_duration_avg',
                  'pages_per_session',
                  'discount_usage_rate',
                  'email_open_rate',
                  'payment_method_diversity'
                  ] #indicates the columns which needs to be filled with 0

df[zero_fill_cols] = df[zero_fill_cols].fillna(0) #fills the missing values with 0

print(df.isnull().sum())
```

```python
#Time based columns 
df['days_since_last_purchase'] = df['days_since_last_purchase'].fillna(df['days_since_last_purchase'].max())

print(df.isnull().sum())
```

```python
#Financial columns
financial_cols = ['average_order_value',
                  'lifetime_value',
                  'credit_balance'
                  ]

df[financial_cols] = df[financial_cols].fillna(df[financial_cols].median())

print(df.isnull().sum())
```

```python
#Rate Columns
df['returns_rate'] = df['returns_rate'].fillna(df['returns_rate'].median())
print(df.isnull().sum())
```

```python
#Demographic columns
df['age'] = df['age'].fillna(df['age'].median())
print(df.isnull().sum())
```

```python
df['signup_quarter'] = df['signup_quarter'].map({
    'Q1': 1, 'Q2': 2, 'Q3': 3, 'Q4': 4
})

df = df[(df['age'] >= 15) & (df['age'] <= 90)]

df['total_purchases'] = df['total_purchases'].clip(lower=0)

df['cart_abandonment_rate'] = df['cart_abandonment_rate'].clip(0, 100)
```

```python
df['churned'] = df['churned'].astype(int)
df['age'] = df['age'].astype(int)
df['days_since_last_purchase'] = df['days_since_last_purchase'].astype(int)
```

```python
#save
df.to_csv("../data/cleaned_dataset.csv", index=False)
```
